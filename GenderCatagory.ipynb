```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame with male data
data = {
    'Count': [1, 2, 3, 4, 5, 6, 7],
    'COUNTRY': ['Eritrea', 'France', 'Ethiopia', 'Germany', 'Brazil', 'USA', 'Uganda'],
    'CATEGORY': ['MALE FROM age of 25_30', 'MALE FROM age of 25_31', 'MALE FROM age of 25_32',
                 'MALE FROM age of 25_33', 'MALE FROM age of 25_34', 'MALE FROM age of 25_35',
                 'MALE FROM age of 25_36'],
    '2015': [7.0777, 5.8058, 7.9311, 8.3185, 8.3617, 6.7542, 7.4190],
    '2016': [6.9512, 5.7434, 8.0659, 8.3632, 8.2306, 6.8261, 7.6032],
    '2017': [6.8734, 5.6638, 8.1614, 8.4013, 8.0985, 6.8814, 7.7962],
    '2018': [6.8741, 5.5555, 8.2249, 8.4343, 7.9739, 6.8892, 7.9724],
    '2019': [6.9436, 5.4262, 8.2862, 8.4590, 7.8722, 6.8384, 8.1382],
    '2020': [7.0787, 5.3241, 8.3438, 8.4672, 7.7983, 6.7439, 8.2892],
    '2021': [7.3563, 5.2540, 8.3906, 8.4586, 7.7489, 6.6274, 8.4094],
    '2022': [7.6992, 5.2058, 8.4362, 8.4342, 7.7154, 6.5215, 8.4828]
}

# Create DataFrame
df_male = pd.DataFrame(data)

# Select specific years (e.g., '2020', '2021', '2022')
selected_years = ['2020', '2021', '2022']

# Filter the data for the selected years
male_data_selected_years = df_male[['COUNTRY'] + selected_years]

# Plotting bar chart for male categories in selected years
plt.figure(figsize=(12, 6))
for year in selected_years:
    plt.bar(male_data_selected_years['COUNTRY'], male_data_selected_years[year], alpha=0.7, label=year)

plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Count of Male Categories Across Selected Years')
plt.legend(title='Year', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
